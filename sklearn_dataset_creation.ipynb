```python
from sklearn.datasets import make_blobs
from sklearn.datasets import make_regression
from sklearn.datasets import make_classification
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=100, centers=3, n_features=2,random_state=0,cluster_std=0.6)
print(X.shape)
X=pd.DataFrame(X)
print(X.head())
plt.scatter(X[0],X[1])
plt.title("Clustering 3 CLusters plot")
plt.show()
```

```python
X,y = make_classification( n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, n_classes=3)
print(X.shape)
X=pd.DataFrame(X)
y=pd.DataFrame(y)
print(X.head(),y.head())
plt.scatter(X[0],y[0])
plt.title("Classification multi category plot")
plt.show()
```

```python
X,y = make_regression(n_samples=100, n_features=1,n_informative=10)
print(X.shape)
X=pd.DataFrame(X)
plt.subplot(1,2,1)
plt.scatter(X,y)
plt.title("Without noise")
#with noise
X,y = make_regression(n_samples=100, n_features=1,n_informative=10,noise=20)
X=pd.DataFrame(X)
plt.subplot(1,2,2)
plt.scatter(X,y)
plt.title("With noise")
plt.show()
```
